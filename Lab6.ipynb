```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import models, layers
```

```python
# load the dataset
df = pd.read_csv('house_data.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
# normalize the data
# a bit crude, but as long as we don't pick "nuisance" columns will be fine
df = (df - df.mean()) / df.std()
```

```python
df.head()
```

```python
df.dtypes
```

# Changes

split the dataset into training and testing sets
start with two input features




# Increased the number of inputs by adding all except id and dates

```python

train_data, test_data, train_targets, test_targets = train_test_split(
    df[[ 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']], df[['price']], test_size=0.2)
```

   .

# Used Relu And Sigmoid For The Activation

Define a neural network with one hidden layer


```python
model = models.Sequential()
```

.

# Changed input shape to match number of inputs in the tests data

```python

model.add(layers.Dense(128, activation='relu', input_shape=(18,)))
model.add(layers.Dense(64, activation='sigmoid'))
model.add(layers.Dense(64, activation='sigmoid'))
model.add(layers.Dense(128, activation='relu'))


```

# Add only one output

```python
model.add(layers.Dense(1))
```

# Compiling the model

```python

model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
```

   .

# Training the model

```python
# train the model
history = model.fit(train_data, train_targets, epochs=300)
```

# Evaluation the performance of the model

```python

test_loss, test_mae = model.evaluate(test_data, test_targets)
```
