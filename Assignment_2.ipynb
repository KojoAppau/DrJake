```python
grade_level_mapping ={'KF':'4','PK':'7','OS':'9'}
df['Student Grade Level'] = df['Student Grade Level'].replace(grade_level_mapping)
df['Student Grade Level'] = pd.to_numeric(df['Student Grade Level'],errors = 'coerce')
```

# 2015 Subset of The DataFrame

# Predictive Model

```python
import seaborn as sns
sns.pairplot(df2_2015, vars=['Discipline Infraction Code', 'Discipline Response Code', 'District Code'], hue='Student Grade Level')

```

```python
from sklearn.model_selection import train_test_split
X = df2_2015[['Discipline Infraction Code', 'Discipline Response Code', 'District Code']]
y = df2_2015['Student Grade Level']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
importance = pd.DataFrame({'feature': X.columns, 'importance': dt.feature_importances_})
importance = importance.sort_values('importance', ascending=False)
print(importance)
```

```python
new_data = pd.DataFrame({'Discipline Infraction Code': [1, 2, 3], 'Discipline Response Code': [4, 5, 6], 'District Code': [7, 8, 9]})
y_new = dt.predict(new_data)
print(y_new)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred, labels=dt.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=dt.classes_)
print('Prediction on test data:\n')
disp.plot()
plt.show()
```

```python
!pip install shap
import shap
explainer = shap.Explainer(dt.predict, X_train)
shap_values = explainer(X_test)
shap.summary_plot(shap_values, X_test, plot_type='bar')
```

```python
from sklearn import tree
#tree.plot_tree(dt);
plt.figure(figsize=(20,20))  # set plot size (denoted in inches)
tree.plot_tree(dt, fontsize=20)
plt.show()

```
